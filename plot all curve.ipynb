```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
curve_rhc=pd.read_csv("rhc.csv",names=["fitness"])
curve_sa=pd.read_csv("sa.csv",names=["fitness"])
curve_ga=pd.read_csv("ga.csv",names=["fitness"])
```

```python
curve_rhc
```

```python
x=np.arange(1,10001)
```

```python
x
```

```python
def plot_all(curve_rhc,curve_sa,curve_ga):
    plt.figure(1)
    plt.plot(x,curve_rhc,markersize=1,color='red',label="Random Hill Climbing")
    plt.plot(x,curve_sa,markersize=1,color='green',label="Simulated Annealing")
    plt.plot(x,curve_ga,markersize=1,color='blue',label="Genetic Algorithm")
    plt.xlabel("Iterations")
    plt.ylabel("Loss value")
    plt.legend()
    plt.title("Loss function value vs Iterations for Neural Network")
    plt.show()
```
