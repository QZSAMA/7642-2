```python
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt
```

```python
algorithms = {
    'Randomized Hill Climbing': mlrose.random_hill_climb,
    'Simulated Annealing': mlrose.simulated_annealing,
    'Genetic Algorithm': mlrose.genetic_alg,
    'MIMIC':mlrose.mimic
}
```

```python
def run_problem(problem,gtid=903789757,num_runs=30,max_attempts=100,max_iters=500):
    results = []
    fitness_curves=[]
    for _ in range(num_runs):
        best_state, best_fitness, fitness_curve = mlrose.random_hill_climb(problem, max_attempts=max_attempts, max_iters=max_iters, restarts=10,curve=True )
        # results.append(best_fitness)
        fitness_values = fill_tail(fitness_curve[:, 0],max_iters)
        # print(len(fitness_values))
        fitness_curves.append(fitness_values)
    return fitness_curves
```

```python
def fill_tail(fitness_values,count):
    current_size=len(fitness_values)
    if current_size<count:
        return np.array(list(fitness_values)+[fitness_values[-1]]*(count-current_size))
    else:
        return fitness_values
        
```

```python
problem1=mlrose.DiscreteOpt(length=8, fitness_fn = mlrose.Queens(), maximize=False, max_val=8)
```

```python
fitness_curves=run_problem(problem1)
```
