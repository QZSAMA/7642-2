```python
import numpy as np
import mlrose_hiive as mlrose

# 4 peak Fitness Function
def four_peaks_fitness(state, t):
    n = len(state)
    ones_at_start = 0
    for i in range(n):
        if state[i] == 1:
            ones_at_start += 1
        else:
            break

    zeros_at_end = 0
    for i in range(n-1, -1, -1):
        if state[i] == 0:
            zeros_at_end += 1
        else:
            break

    if ones_at_start > t and zeros_at_end > t:
        return max(ones_at_start, zeros_at_end) + n
    else:
        return max(ones_at_start, zeros_at_end)
```

```python
t = 10
fitness = mlrose.CustomFitness(lambda state: four_peaks_fitness(state, t))
problem = mlrose.DiscreteOpt(length=30, fitness_fn=fitness, maximize=True, max_val=2)
```

```python
# 随机爬山算法
best_state_rhc, best_fitness_rhc, _ = mlrose.random_hill_climb(problem, max_attempts=100, max_iters=1000, restarts=10)
print("Random Hill Climb: Best Fitness =", best_fitness_rhc)
```

```python
schedule = mlrose.ExpDecay(init_temp=100, exp_const=0.005, min_temp=1)
best_state_sa, best_fitness_sa, _ = mlrose.simulated_annealing(problem, schedule=schedule, max_attempts=100, max_iters=1000)
print("Simulated Annealing: Best Fitness =", best_fitness_sa)
```

```python
best_state_ga, best_fitness_ga, _ = mlrose.genetic_alg(problem, pop_size=200, mutation_prob=0.1, max_attempts=100, max_iters=1000)
print("Genetic Algorithm: Best Fitness =", best_fitness_ga)
```

```python
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt

# 定义n皇后问题的适应度函数
fitness = mlrose.Queens()

# 创建优化问题对象
problem = mlrose.DiscreteOpt(length=8, fitness_fn=fitness, maximize=False, max_val=8)

# 定义要测试的算法列表
algorithms = {
    'Randomized Hill Climbing': mlrose.random_hill_climb,
    'Simulated Annealing': mlrose.simulated_annealing,
    'Genetic Algorithm': mlrose.genetic_alg
}

# 记录结果
results = {}

# 运行每个算法
for name, algorithm in algorithms.items():
    best_state, best_fitness, fitness_curve = algorithm(problem, max_attempts=max_attempts, max_iters=max_iters, curve=True)
    results[name] = (best_state, best_fitness, fitness_curve)
    print(f"{name} - Best state found: {best_state}, Best fitness value (conflicts): {best_fitness}")

# 绘制每个算法的收敛曲线
plt.figure()
for name, (best_state, best_fitness, fitness_curve) in results.items():
    fitness_values = fitness_curve[:, 0]
    plt.plot(fitness_values, label=name)

# 添加图例和标签
plt.title('Convergence Curves for N-Queens Problem')
plt.xlabel('Iterations')
plt.ylabel('Fitness Value (Number of Conflicts)')
plt.legend()
plt.show()
```

```python
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt

# Define the fitness function for k-coloring problem
class KColorFitness:
    def __init__(self, edges, k):
        self.edges = edges
        self.k = k

    def evaluate(self, state):
        # state is a list of colors assigned to each node
        fitness = 0
        for edge in self.edges:
            if state[edge[0]] != state[edge[1]]:
                fitness += 1
        return fitness

# Define the problem
edges = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4)]
k = 3
fitness_fn = mlrose.CustomFitness(lambda state: KColorFitness(edges, k).evaluate(state))
problem = mlrose.DiscreteOpt(length=len(edges), fitness_fn=fitness_fn, maximize=True, max_val=k)

# Solve the problem using Genetic Algorithm
best_state, best_fitness, fitness_curve = mlrose.genetic_alg(problem, max_iters=1000, curve=True)

# Print the results
print("Best state found:", best_state)
print("Best fitness found:", best_fitness)

# Plot the fitness curve
plt.plot(fitness_curve)
plt.title('Fitness Curve for Genetic Algorithm on k-Coloring Problem')
plt.xlabel('Iterations')
plt.ylabel('Fitness Value')
plt.show()
```

```python
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt

# Define the Four Peaks fitness function
fitness_fn = mlrose.FourPeaks(t_pct=0.15)
problem = mlrose.DiscreteOpt(length=20, fitness_fn=fitness_fn, maximize=True)

# Solve the problem using Simulated Annealing
best_state, best_fitness, fitness_curve = mlrose.simulated_annealing(problem, max_attempts=100, max_iters=1000, curve=True)

# Print the results
print("Best state found:", best_state)
print("Best fitness found:", best_fitness)

# Plot the fitness curve
plt.plot(fitness_curve[:, 0])
plt.title('Fitness Curve for Simulated Annealing on Four Peaks Problem')
plt.xlabel('Iterations')
plt.ylabel('Fitness Value')
plt.show()
```
